```python
import numpy as np

D_e = 5.
alpha = 0.5
r0 = 3.

def Morse(r, D_e, alpha, r0):
    return D_e * ((1. - np.exp(-alpha * (r - r0)))**2. - 1.)
```

```python
N = 10001
a = 0.
b = 30
```

```python
import matplotlib.pyplot as plt

x = np.linspace(a, b, N)
h = x[1]-x[0]

v = Morse(x, D_e, alpha, r0)

plt.plot(x,v)
plt.show()
```

```python
D = np.zeros(N)
D[:] = 2./h**2 + v[:]

E = np.zeros(N-1)
E[:] = -1/h**2
```

```python
from scipy.linalg import eigh_tridiagonal

# calcola solo gli autovalori negativi
eigenvals, eigenvecs = eigh_tridiagonal(D, E, select='v', select_range=(-D_e, 0))

eigenvals = eigenvals/2
eigenvecs[:,:] = eigenvecs[:,:] * h**-0.5

print('%i autovalori calcolati' %len(eigenvals))
```

```python
fig, ax = plt.subplots()

ax.plot(x, eigenvecs)
plt.show()
```
