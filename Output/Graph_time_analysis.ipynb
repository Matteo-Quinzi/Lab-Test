```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('seaborn-notebook')

data = np.loadtxt("time_analysis_waves.txt")

L = data[:,0]
x_time = data[:,1]
x_time_err = data[:,2]
k_time = data[:,3]
k_time_err = data[:,4]

fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
plt.xlabel("interval length")
axs[0].set_yscale("log")
axs[0].set_ylabel("CPU time")
axs[0].errorbar(L, x_time, yerr=x_time_err, fmt='o')
axs[0].plot(L, x_time)

axs[1].set_yscale("log")
axs[1].set_ylabel("CPU time")
axs[1].errorbar(L, k_time, yerr=k_time_err, fmt='o')
axs[1].plot(L, k_time)

plt.savefig("CPU time")
plt.show()
```
